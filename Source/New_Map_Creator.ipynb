```python
# Import the pandas library
import pandas as pd
# import the library
import folium
#load the data
data = pd.read_csv('countries.csv')
```

```python
#different map themes at: https://deparkes.co.uk/2016/06/10/folium-map-tiles/

# Make an empty map
m = folium.Map(location=[52.132633,5.291266], tiles="stamentoner", zoom_start=2)

# Show the map
#m
```

```python


# Make a data frame with dots to show on the map

for i, naam in enumerate(data['country']):
    try:
        float(data['latitude'][i])
        boolean = True
    except:
        boolean = False
    if not boolean:
        data['Latitude'][i] = 0
        data['Longitude'][i] = 0

        
data = data.dropna()
```

```python
# add marker one by one on the map
for i in range(0,len(data)):
    title = str(data.iloc[i]['country']) + ", " + str(data.iloc[i]['count'])
    folium.Marker(
      location=[data.iloc[i]['latitude'], data.iloc[i]['longitude']],
      popup=title,
    ).add_to(m)

# Show the map again
m
```
